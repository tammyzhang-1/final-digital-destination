```python
import pandas as pd
import numpy as np
import json
```

```python
devices = pd.read_csv("devices.csv")
devices["certification"] = devices["certification"].replace({np.nan:"None"})
devices.to_json("devices.json", orient = "records" )
```

```python
nodes = pd.read_csv("nodes.csv")
```

```python
finalLocationList = {}

def createLocationList(df):
    #print(strip(df['name'].values))
    locationList = df['name'].values.tolist()
    locationListStripped = [location.strip() for location in locationList]
    finalLocationList[str(df['device_id'].iloc[0])] = locationListStripped
    
    
    #finalLocationList.append({df['device_id'].iloc[0] : locationListStripped}) 
nodes.groupby("device_id", as_index=False).apply(createLocationList)

finalLocationList
```

```python
with open("nodesList.json", "w") as file:
    json.dump(finalLocationList, file)
```

```python
feature_collection_template = {"type": "FeatureCollection" }
def create_feature(row):
    coordinates = json.loads(row["coordinates"])
    return {
        "type": "Feature",
        "geometry": {"type": "Point", "coordinates": coordinates},
        "properties": {
            "id": row["id"],
            "device_id": row["device_id"],
            "name": row["name"],
            "device_type": devices.loc[devices['device_id'] == row["device_id"], 'device_type'].values[0],
            "precise": row["precise"]
        }
    }
```

```python
feature_collection = nodes.apply(create_feature, axis=1).tolist()
feature_collection_template["features"] = feature_collection
feature_json = feature_collection_template
with open("nodes.geojson", "w") as f:
    json.dump(feature_json, f)
```

```python
feature_collection_template2 = {"type": "FeatureCollection" }

def generateEdges(df):
    return df.iloc[:-1].apply(create_feature2, axis=1)

def create_feature2(row):
    next_row_index = row["id"] + 1
    
    if next_row_index < len(nodes) + 1:
        next_row = nodes.iloc[next_row_index - 1]
        next_coordinates = json.loads(next_row["coordinates"])
        coordinates = json.loads(row["coordinates"])
        
        # cross if coords are on separate sides of the world
        if coordinates[0] * next_coordinates[0] < 0:
            # but don't cross if leaving from east coast of US
            if -100 < coordinates[0] < -66:
                cross = 0
            else:
                cross = 1
        else: 
            cross = 0
        
        return { "type": "Feature",
          "geometry": {
            "type": "LineString",
            "coordinates": [coordinates, next_coordinates]
            },
          "properties": {
            "device_id": row["device_id"],
            "source_id": int(row["id"]),
            "target_id": int(next_row["id"]),
            "cross": cross,
            "device_type": devices.loc[devices['device_id'] == row["device_id"], 'device_type'].values[0]
            }
          }
    return None
```

```python
edges_unfiltered = nodes.groupby("device_id", as_index=False).apply(generateEdges).tolist()
```

```python
edges = [i for i in edges_unfiltered if i is not None]
```
